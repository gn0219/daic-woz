```python
from sklearn.preprocessing import StandardScaler

# Drop unnecessary columns
columns_to_drop = ['PHQ8_Binary', 'PHQ8_Score', 'Participant_ID', 'Gender', 'wav_path', 'start', 'end']
X = df.drop(columns=columns_to_drop)
y = df['PHQ8_Binary']

# Split the dataset based on the 'Split' column
X_train = X[df['Split'] == 'train'].drop(columns=['Split'])
y_train = y[df['Split'] == 'train']

X_dev = X[df['Split'] == 'dev'].drop(columns=['Split'])
y_dev = y[df['Split'] == 'dev']

X_test = X[df['Split'] == 'test'].drop(columns=['Split'])
y_test = y[df['Split'] == 'test']

# Apply StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_dev_scaled = scaler.transform(X_dev)
X_test_scaled = scaler.transform(X_test)
```

```python
# Check for NaN or infinite values in the scaled datasets
print("Checking for NaN or infinite values in scaled datasets:")

print("X_train_scaled contains NaN:", np.isnan(X_train_scaled).any())
print("X_train_scaled contains infinite values:", not np.isfinite(X_train_scaled).all())

print("X_dev_scaled contains NaN:", np.isnan(X_dev_scaled).any())
print("X_dev_scaled contains infinite values:", not np.isfinite(X_dev_scaled).all())

print("X_test_scaled contains NaN:", np.isnan(X_test_scaled).any())
print("X_test_scaled contains infinite values:", not np.isfinite(X_test_scaled).all())
```

```python
# Replace NaN values with the mean of each column
X_dev_scaled = np.nan_to_num(X_dev_scaled, nan=np.nanmean(X_dev_scaled, axis=0))
X_test_scaled = np.nan_to_num(X_test_scaled, nan=np.nanmean(X_test_scaled, axis=0))

# Replace infinite values with large finite numbers
X_dev_scaled = np.nan_to_num(X_dev_scaled, posinf=np.finfo(np.float64).max, neginf=np.finfo(np.float64).min)
X_test_scaled = np.nan_to_num(X_test_scaled, posinf=np.finfo(np.float64).max, neginf=np.finfo(np.float64).min)

# Verify if there are still NaN or infinite values
print("X_dev_scaled contains NaN:", np.isnan(X_dev_scaled).any())
print("X_dev_scaled contains infinite values:", not np.isfinite(X_dev_scaled).all())

print("X_test_scaled contains NaN:", np.isnan(X_test_scaled).any())
print("X_test_scaled contains infinite values:", not np.isfinite(X_test_scaled).all())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, f1_score

def train_and_evaluate_models(X_train, y_train, X_dev, y_dev, X_test, y_test):
    # Initialize models
    models = {
        'SVM': SVC(kernel='linear', random_state=42, probability=True),
        'KNN': KNeighborsClassifier(n_neighbors=5),
        'RF': RandomForestClassifier(n_estimators=100, random_state=42)
    }

    # Dictionary to store results
    results = {}

    # Train and evaluate each model
    for model_name, model in models.items():
        print(f"Training {model_name}...")
        model.fit(X_train, y_train)

        # Evaluate on train, dev, and test sets
        train_preds = model.predict(X_train)
        dev_preds = model.predict(X_dev)
        test_preds = model.predict(X_test)

        train_probs = model.predict_proba(X_train)[:, 1] if hasattr(model, "predict_proba") else None
        dev_probs = model.predict_proba(X_dev)[:, 1] if hasattr(model, "predict_proba") else None
        test_probs = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

        train_acc = accuracy_score(y_train, train_preds)
        dev_acc = accuracy_score(y_dev, dev_preds)
        test_acc = accuracy_score(y_test, test_preds)

        train_f1 = f1_score(y_train, train_preds, average='macro')
        dev_f1 = f1_score(y_dev, dev_preds, average='macro')
        test_f1 = f1_score(y_test, test_preds, average='macro')

        train_auroc = roc_auc_score(y_train, train_probs) if train_probs is not None else None
        dev_auroc = roc_auc_score(y_dev, dev_probs) if dev_probs is not None else None
        test_auroc = roc_auc_score(y_test, test_probs) if test_probs is not None else None

        # Store results
        results[model_name] = {
            'Train Accuracy': train_acc,
            'Dev Accuracy': dev_acc,
            'Test Accuracy': test_acc,
            'Train F1': train_f1,
            'Dev F1': dev_f1,
            'Test F1': test_f1,
            'Train AUROC': train_auroc,
            'Dev AUROC': dev_auroc,
            'Test AUROC': test_auroc
        }

        # Print classification report for test set
        print(f"\nClassification Report for {model_name} on Test Set:")
        print(classification_report(y_test, test_preds))

    # Convert results to DataFrame
    results_df = pd.DataFrame(results).T

    # Print results
    print("\nResults in Dictionary Format:")
    print(results)
    print("\nResults in DataFrame Format:")
    print(results_df)

    return results, results_df

# Call the function with the scaled data
results, results_df = train_and_evaluate_models(X_train_scaled, y_train, X_dev_scaled, y_dev, X_test_scaled, y_test)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

# Perform feature selection using Random Forest
feature_selector = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42), threshold="median")
X_train_selected = feature_selector.fit_transform(X_train_scaled, y_train)
X_dev_selected = feature_selector.transform(X_dev_scaled)
X_test_selected = feature_selector.transform(X_test_scaled)

# Get the selected feature indices
selected_features = feature_selector.get_support(indices=True)
print("Selected feature indices:", selected_features)

# Run ML models with the selected features
results_selected, results_df_selected = train_and_evaluate_models(
    X_train_selected, y_train, X_dev_selected, y_dev, X_test_selected, y_test
)

# Display the results
print("Results with selected features:")
print(results_df_selected)
```
